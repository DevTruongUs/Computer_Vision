```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from PIL import Image

# Doc anh
image = Image.open('C:/Users/Admin/OneDrive - VNU-HCMUS/Documents/src.cv/img/01/animal.jpg')

# Chuyen sang anh xam
gray_image = image.convert('L')
image_array = np.array(gray_image)

# Chuyen sang anh nhi phan
binary_image = 1*(image_array<200)

# Thuc hien openning de loai bo nhieu va noi cac doi tuong
opened_image = ndimage.binary_opening(binary_image, structure=np.ones((9, 5)))

# Danh nhan
labeled_image, num_features = ndimage.label(opened_image)

# Tao luoi subplot voi 3 dong va 1 cot
fig, axs = plt.subplots(4, 1, figsize=(50, 50))

plt.subplot(4,1,1)
plt.imshow(gray_image, cmap='gray')
plt.title("Anh xam")

plt.subplot(4,1,2)
plt.imshow(binary_image, cmap='gray')
plt.title("Anh nhi phan")

plt.subplot(4,1,3)
plt.imshow(opened_image, cmap='gray')
plt.title("Anh sau opening")

plt.subplot(4,1,4)
plt.imshow(labeled_image, cmap='nipy_spectral')
plt.title(f"Anh da danh nhan ({num_features} doi tuong)")

plt.show()
```
