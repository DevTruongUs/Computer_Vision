```python
import matplotlib as plt
import matplotlib.pyplot as plt
```

```python
# Ve bieu do: Line Chart 
x = [1, 2, 3, 4, 5]
y = [10, 16, 20 , 30, 25]

x1 = [1, 2, 3, 4, 5]
y1 = [20, 15, 10 , 9, 30]

plt.plot(x, y, 'bo-', label='Du lieu')
plt.plot(x1, y1, 'ro-', label='Du lieu-1')
plt.legend()
plt.xlabel('Truc x')
plt.ylabel('Truc y')
plt.show()
```

```python
# Ve bieu do cot Bar Chart
categories = ['A', 'B', 'C', 'D', 'E']
values = [15, 10, 30, 25, 15]

plt.bar(categories, values, color = 'g', alpha=0.6)
plt.xlabel('Danh Muc')
plt.ylabel('Gia Tri')
plt.show()
```

```python
# Ve bieu do hinh tron Pie Chart
my_labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
my_colors = ['green', 'blue', 'red', 'yellow']

plt.pie(sizes, labels=my_labels, colors = my_colors, autopct='%1.1f%%')
plt.title('Bieu do hinh tron')
plt.show()
```

```python
# Bieu do Phan tan Scatter Chart
x=[1, 2, 3, 4, 5]
y=[10, 16, 20, 30, 25]

plt.scatter(x, y, c='r', marker='o', label='Du lieu')
plt.title('Bieu do phan tan vi du')
plt.xlabel('Truc x')
plt.ylabel('Truc y')
plt.legend()
plt.show()
```

```python
# Bieu do Hop 
data = [18, 15, 23, 58, 40, 70, 30, 55, 21, 30]

plt.boxplot(data)
plt.title('Bieu do hop vi du')
plt.ylabel('Gia tri')
plt.show()
```

```python
# Bieu do Violin
data = [15, 10, 22, 30, 35, 45, 50, 55, 65]

plt.violinplot(data)
plt.title('Bieu do Violin vi du')
plt.ylabel('Gia Tri')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 4, 3, 7, 6]

N = len(categories)
values += values[:1]

angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]

plt.polar(angles, values, 'o--', linewidth=2)
plt.fill(angles, values, 'b', alpha=0.1)
plt.title('Bieu do radar vi du')
plt.show()
```
