```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Duong dan anh
my_path = 'C:/Users/Admin/OneDrive - VNU-HCMUS/Documents/src.cv/img/01/1.jpg'

# Doc anh
img = Image.open(my_path)
plt.imshow(img)
```

```python
# Chuyen sang anh xam
img_gray = img.convert('L')
plt.imshow(img_gray, cmap='gray')
```

```python
img_array = np.array(img_gray)
print(img_array)
```

```python
# Tao mot hinh anh moi
plt.figure()
plt.hist(img_array.flatten(), bins=128)
plt.title('Bieu do Histogram cua anh')
plt.xlabel('Gia tri cua pixel')
plt.ylabel('So luong')
plt.show()
```

```python
#lam noi bat duong bien
plt.figure()
plt.gray()
plt.contour(img_gray, origin="image")
plt.show()
```
